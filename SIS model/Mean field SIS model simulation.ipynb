```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter, AutoMinorLocator)
import seaborn as sns
```

```python
def euler(x, beta=1, gamma=1):
    h=0.1
    y=[x[0]]
    z=[x[1]]
    for n in range(0, 1000):
        tempy = y[n]
        tempz = z[n]
        a=np.array(tempy) + (gamma*np.array(tempz) - beta*np.array(tempy)*np.array(tempz))*h
        b=np.array(tempz) + (beta*np.array(tempy)*np.array(tempz) - gamma*np.array(tempz))*h
        a=a.tolist()
        b=b.tolist()
        y.append(a)
        z.append(b)
    return z
```

```python
yy=euler([0.999, 0.001], beta=1.3, gamma=1)
plt.plot(yy)
plt.xlabel('iteration steps')
plt.ylabel('Fraction infected')
plt.show()
```

```python
xx=np.linspace(0, 4, 30)
yy=[]
for i in range(0,len(xx)):
    yy.append(euler([0.999, 0.001], xx[i], gamma=1)[-1])

plt.plot(xx, yy)
plt.xlabel('infectious rate')
plt.ylabel('Fraction infected')
plt.show()
```
